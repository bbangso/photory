```python
from util.data import sentencePieceTokenizer, toString
import re
```

```python
sentencepieceTokenizer= sentencePieceTokenizer()
```

```python
file = open('Dataset/Grimm_Original_UTF.txt', 'r', encoding='utf-8')
```

```python
untokenized_file = open('Dataset/untokenized_Grimm.txt', 'w', encoding='utf-8')
tokenized_file = open('Dataset/tokenized_Grimm.txt', 'w', encoding='utf-8')
```

```python
untokenized = ""
tokenized = ""
data_length = 0
count = 1
WRITE_MODE = False
PRE_WRITE_MODE =False
FIRST_SENTENCE_CHECK = False
NEXT_LINE_CHECK = False
```

```python
str_temp = ''
print(str_temp.isalpha())
```

```python
while True:
    line = file.readline()
    
    if not line:
      untokenized_file.write("<s>" + untokenized+"</s>"+'\n')
      tokenized_file.write(tokenized)
      break
        
    if NEXT_LINE_CHECK == True :
        NEXT_LINE_CHECK = False
        continue
    line = line.replace('‘','').replace('’','')
    if PRE_WRITE_MODE == True :
        if ('(본문시작)' in line or '(분문 시작)' in line or '(본문 시작)' in line) :
            PRE_WRITE_MODE = False
            WRITE_MODE = True
            continue
            
        elif FIRST_SENTENCE_CHECK == True :
            if not (line == '' or line == '\n' or line == ' ' or line == '\t') :
                #print(str(line))
                FIRST_SENTENCE_CHECK = False
                if line[0].isalpha() == True or line[0] == "“" or line[0].isdigit() == True :
                    PRE_WRITE_MODE = False
                    WRITE_MODE = True
                else : 
                    continue
            else :
                continue
            
                
        else :
            continue
    
    
        
    line = line.replace('`','').replace('\'', '')
    if '(끝)' not in line and '(동화 끝)' not in line :
        line = re.sub(r'\([^)]*\)', '', line)
    tokenized_line = sentencepieceTokenizer(line)
    #print(line)
    
    
    if data_length+len(tokenized_line)+2 >= 1022: # bos와 eos 토큰 갯수 고려 +2
      #print(data_length+len(tokenized_line)+2)
      #print('tokenized_line : ', len(tokenized_line))
      #print('untokenized_line : ', len(line))
      
      untokenized_file.write("<s>"+untokenized+"</s>"+'\n')
      tokenized_file.write("<s>"+tokenized+ "</s>"+'\n')
        
      tokenized = ""
      untokenized = ""
      data_length = 0
      
    
    if '(끝)' in line or '(동화 끝)' in line :
        if untokenized != "" :
            untokenized_file.write("<s>"+untokenized+"</s>"+'\n')
            tokenized_file.write("<s>"+tokenized+"</s>"+'\n')
        untokenized = ""
        tokenized = ""
        data_length = 0
        WRITE_MODE = False
        PRE_WRITE_MODE = False
        
    if WRITE_MODE == True :
        if data_length+len(tokenized_line) + 2 > 2044 :
            if untokenized != "" :
                untokenized_file.write("<s>"+untokenized+"</s>"+'\n')
                tokenized_file.write("<s>"+tokenized+"</s>"+'\n')
            untokenized = line[:-1]
            tokenized = toString(tokenized_line)
            data_length = len(tokenized_line) + 2
        else :
            untokenized = untokenized + line[:-1]
            tokenized = tokenized + toString(tokenized_line)
            data_length = data_length+len(tokenized_line) + 2 # bos와 eos 토큰 갯수 고려 +2
        
    if '{}.'.format(count) in line and WRITE_MODE == False :
        PRE_WRITE_MODE = True
        FIRST_SENTENCE_CHECK = True
        NEXT_LINE_CHECK = True
        count += 1
        print(line)
        
    
print(str(count) + '<end>')

file.close()
untokenized_file.close()
tokenized_file.close()
```
